```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


def drawElectrodeGrid(cols, rows, gridWidth, gridHeight, gridname, xlabel, ylabel, electrodeLabel, electrodeLabelStartsAt): 
    gridOriginX = 0
    gridOriginY = 0

    # Grid parameters for the 4x2 points in front of the rectangle

    spacing_x = gridWidth / (cols-1)  # spread across the rectangle width
    spacing_y = gridHeight / (rows-1)  # vertical spacing in front of the rectangle


    xs = []
    ys = []
    for r in range(rows):
        for c in range(cols):
            x = gridOriginX + c * spacing_x
            y = gridOriginY + r * spacing_y
            xs.append(x)
            ys.append(y)

    rectangleMarginX = gridWidth * 0.25
    rectangleMarginY = gridHeight * 0.25

    # Rectangle parameters (x, y, width, height)
    rect_x = gridOriginX - rectangleMarginX
    rect_y = gridOriginY - rectangleMarginY
    rect_w = gridOriginX + gridWidth + 2*rectangleMarginX
    rect_h = gridOriginY + gridHeight + 2*rectangleMarginY

    fig, ax = plt.subplots(figsize=(12, 7))

    rect = Rectangle((rect_x, rect_y), rect_w, rect_h, fill=False, linewidth=4)
    ax.add_patch(rect)

    ax.plot(xs, ys, marker='o', markersize=15) # points 
    ax.plot(xs, ys, linewidth=5)  # line connecting all points

    for i, (x, y) in enumerate(zip(xs, ys), start=1):
        if i == 1 or i == cols*rows:
            electrodeLabelNumber = i+electrodeLabelStartsAt-1
            ax.text(x, y + 1, f"{electrodeLabel}{electrodeLabelNumber:03d}", ha='center', va='center', fontsize=15) # label electrodes

    ax.text(gridOriginX + 0.5*gridWidth, gridOriginY + gridHeight + 0.5 * rectangleMarginY, gridname, horizontalalignment='center', verticalalignment='center', fontsize=18)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    ax.set_aspect('equal', adjustable='box')
    plt.show()
```

```python
# HDsEMGs
drawElectrodeGrid(cols=4, rows=2, gridWidth = 30, gridHeight = 10, gridname="Grid1", xlabel="Grid1CoordSys X", ylabel="Grid1CoordSys Y", electrodeLabel="E", electrodeLabelStartsAt=1)
drawElectrodeGrid(cols=2, rows=4, gridWidth = 5, gridHeight = 15, gridname="Grid2", xlabel="Grid2CoordSys X", ylabel="Grid2CoordSys Y", electrodeLabel="E", electrodeLabelStartsAt=9)
drawElectrodeGrid(cols=3, rows=3, gridWidth = 20, gridHeight = 20, gridname="Grid3", xlabel="Grid3CoordSys X", ylabel="Grid3CoordSys Y", electrodeLabel="E", electrodeLabelStartsAt=17)
# intra Grids 
drawElectrodeGrid(cols=2, rows=4, gridWidth = 10, gridHeight = 30, gridname="intraGrid1", xlabel="intraGrid1CoordSys X", ylabel="superficial <--> deep", electrodeLabel="iE", electrodeLabelStartsAt=1)
drawElectrodeGrid(cols=4, rows=2, gridWidth = 30, gridHeight = 10,gridname="intraGrid2", xlabel="superficial <--> deep", ylabel="intraGrid2CoordSys Y", electrodeLabel="iE", electrodeLabelStartsAt=9)
```
